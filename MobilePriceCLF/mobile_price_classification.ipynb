**Data Split**

```python
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
X, y = df.drop('price_range', axis=1), df['price_range']
X, y = shuffle(X, y, random_state=15)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
X_train.shape, X_test.shape
```

# **Model Building**

```python
# I'll start with RF
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, \
f1_score, precision_score, recall_score
```

```python
forest = RandomForestClassifier()  # rf out of the box
forest.fit(X_train, y_train)
```

```python
# Accuracy Metric

train_pred = forest.predict(X_train)
test_pred = forest.predict(X_test)

print(f'Accuracy on Train: ', accuracy_score(y_train, train_pred))
print(f'Accuracy on Holdout: ', accuracy_score(y_test, test_pred))
```

```python
# good, but I'm aware of 1.0 accuracy on training
```

```python
print(f'F1 on Train: ', f1_score(y_train, train_pred, average='macro'))
print(f'F1 on Holdout: ', f1_score(y_test, test_pred, average='macro'))
```

```python
print(f'Precision on Holdout: ', precision_score(y_test, test_pred, average='macro'))
print(f'Recall on Holdout: ', recall_score(y_test, test_pred, average='macro'))
```
