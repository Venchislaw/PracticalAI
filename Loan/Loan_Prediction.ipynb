**Okay... We have Binary classification problem** <br>
Possible algorithms (from easiest to most powerful):

3. SVM (must try)
4. Random Forest / XGBoost (idk what to pick)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import RandomOverSampler
```

```python
X = df.drop('Personal Loan', axis=1)
y = df['Personal Loan']
```

```python
ros = RandomOverSampler(random_state=0)
X, y = ros.fit_resample(X, y)
```

```python
X, y = shuffle(X, y)
```

```python
df.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
logistic_regression = LogisticRegression()

cross_val_score(estimator=logistic_regression,
                X=X_train, y=y_train,
                cv=KFold())
```

```python
logistic_regression_cv = cross_val_score(estimator=logistic_regression,
                X=X_train, y=y_train,
                cv=KFold()).mean()

logistic_regression.fit(X_train, y_train)
test_pred = logistic_regression.predict(X_test)
accuracy_score(y_test, test_pred)
```

```python
# KNN test
```

```python
neighbors_classifier = KNeighborsClassifier(n_jobs=-1)

cross_val_score(estimator=neighbors_classifier,
                X=X_train, y=y_train,
                cv=KFold())
```

```python
neighbors_cv = cross_val_score(estimator=neighbors_classifier,
                X=X_train, y=y_train,
                cv=KFold()).mean()

neighbors_classifier.fit(X_train, y_train)
test_pred = neighbors_classifier.predict(X_test)
accuracy_score(y_test, test_pred)
```

```python
np.bincount(test_pred)
```

```python
# random forest time
```

```python
random_forest = RandomForestClassifier(max_depth=10)
random_forest.fit(X_train, y_train)

test_pred = random_forest.predict(X_test)
accuracy_score(y_test, test_pred)
```

```python
# no way...
```

```python
random_forest_cv = cross_val_score(
    estimator=random_forest,
    X=X_train,
    y=y_train,
    cv=KFold()
)

random_forest_cv = random_forest_cv.mean()
```

```python
(random_forest_cv, random_forest_cv.mean())
```

```python
algorithms = ['Log_Regression', 'KNN', 'RandForest']
results = [logistic_regression_cv, neighbors_cv, random_forest_cv]

sns.barplot(x=algorithms, y=results, hue=algorithms, legend=algorithms)
plt.title('Algorithms performance')
plt.legend()
plt.plot();
```
